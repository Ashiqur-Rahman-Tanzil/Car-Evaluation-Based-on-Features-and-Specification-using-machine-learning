```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

# Loading and Reading dataset

```python
data = pd.read_csv("car.data")
print(data.head())  
# To check if our data is loaded correctly
```

```python
le = preprocessing.LabelEncoder()
```

```python
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
predict = "class"
```

```python
X = list(zip(buying, maint, door, persons, lug_boot, safety))  # features
y = list(cls)  # labels
```

# Spliting Train and  Test set

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
model = KNeighborsClassifier(n_neighbors=9)
model.fit(x_train, y_train)
```

# Train Accuracy

```python
acc = model.score(x_train, y_train)
print(acc)
```

# Test Accuracy

```python
acc = model.score(x_test, y_test)
print(acc)
```

# Results with nearest neighbors

```python
predicted = model.predict(x_test)
names = ["1.25", "2.50", "3.75", "5.00"]

for x in range(len(predicted)):
    print("Predicted Grade: ", names[predicted[x]],  "Actual Grade: ", names[y_test[x]])
    n = model.kneighbors([x_test[x]], 5, True)
    print("Neighbors distance and index: ", n)
```
